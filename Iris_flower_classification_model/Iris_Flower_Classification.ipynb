```python
from sklearn.model_selection import train_test_split
x = iris.drop(["species"], axis = 1)
y = iris["species"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

Logistic Regression


```python
LR = LogisticRegression()
LR.fit(x_train, y_train)
prediction = LR.predict(x_test)
print("Accuracy Score: ", metrics.accuracy_score(prediction,y_test))
```

Decision Tree

```python
DT = DecisionTreeClassifier()
DT.fit(x_train, y_train)
prediction = DT.predict(x_test)
print("Accuracy Score: ", metrics.accuracy_score(prediction, y_test))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 20))
plot_tree(DT, filled=True, feature_names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], class_names=iris.species)
plt.title("Decision Tree for Iris Species Classification")
plt.show()
```
